**Importando bibliotecas**

```
from gitly.colab.plot import GitlyPlotter
gitly = GitlyPlotter('git')
gitly.config_render(renderer='colab')
import pandas as pd
import plotly.express as px
from google.colab import drive
drive.mount('/content/drive')
```

**Importando dados**

```
df_compra_itens = pd.read_csv('/content/drive/My Drive/Pós-Graduação/Introdução a Ciência de Dados/olist_order_items_dataset.csv')
df_produtos = pd.read_csv('/content/drive/My Drive/Pós-Graduação/Introdução a Ciência de Dados/olist_products_dataset.csv')
```

**Produtos mais vendidos**

```
df_compra_itens.head()
```

```
df_prod_mais_v = df_compra_itens[['product_id','order_item_id']].groupby(by='product_id').apply(sum)['order_item_id'].sort_values(ascending = False)
```

```
df_prod_mais_v2 = df_prod_mais_v.head()
df_prod_mais_v2
```

```
fig = px.bar(df_prod_mais_v2, x='order_item_id', title='Produtos mais vendidos', labels={'order_item_id': 'Número de vendas', 'index':'Product_id'})
gitly.show( fig )
```

**Produtos com maiores faturamentos**

```
df_compra_itens['Valor_final'] = df_compra_itens.order_item_id * df_compra_itens.price
df_compra_itens.head()
```

```
df_fat_mais = df_compra_itens[['product_id','order_item_id','Valor_final']].groupby(by='product_id').apply(sum)[['Valor_final','order_item_id']].sort_values(by='Valor_final',ascending = False)
df_fat_mais.head()
```

```
df_fat_mais2 = df_fat_mais.head()
```

```
fig = px.bar(df_fat_mais2, y='Valor_final', title='Produtos com maiores faturamentos', labels={'Valor_final': 'Valor', 'index':'Product_id'})
gitly.show( fig )
```

**Venderores com mais vendas**

```
df_vend_mais = df_compra_itens[['seller_id','order_item_id']].groupby(by='seller_id').apply(sum)['order_item_id'].sort_values(ascending = False)
df_vend_mais.head()
```

```
df_vend_mais = df_compra_itens[['seller_id','order_item_id']].groupby(by='seller_id').sum().sort_values(by='order_item_id',ascending = False)
df_vend_mais.head()
```

```
df_vend_mais2 = df_vend_mais.head()
```

```
fig = px.bar(df_vend_mais2, y='order_item_id', title='Vendedores com mais vendas', labels={'order_item_id': 'Qtd. vendida', 'index':'seller_id'})
gitly.show( fig )
```

**Categorias mais vendidas**

```
df_produtos.head()
```

```
df_right = df_produtos[['product_id','product_category_name']]
df_right
```

```
df_left = df_compra_itens[['product_id','order_item_id']]
df_left
```

```
df_categoria = pd.merge(df_left, df_right, how='left',on='product_id')
df_categoria
```

```
df_cat_mais = df_categoria[['product_category_name','order_item_id']].groupby(by='product_category_name').apply(sum).sort_values(by='order_item_id', ascending = False)
df_cat_mais.head(5)
```

```
df_plot = df_cat_mais.sort_values(by='order_item_id', ascending=False).head(5)
fig = px.bar(df_plot, x='order_item_id', title='Categorias mais vendidas', labels={'order_item_id': 'Qtd. Vendidas', 'index':'Categoria'}, orientation='h')
gitly.show( fig )
```
